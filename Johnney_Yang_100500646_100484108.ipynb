```python
def monte_carlo(mu, sigma, no_days): # Defines the function of monte carlo with 3 paramaters of mu, sigma and no_days
    P = np.zeros(no_days) #creates an array with number of days as zero and sets day 0 initial price as 1250
    P[0] = 1250.0

    Z = np.random.normal(mu, sigma, no_days) #Creates an array which generates randomly distributed numbers with a mean of mu and standard deviation of sigma

    R = Z + 1 # Calculates multipliers

    for i in range(1, no_days): # Loops through the number of days and -1 each time
        P[i] = P[i-1] * R[i-1] # Updates price of i with the previous day by the corresponding multiplier       
    return P #Returns the array with the simulated price
```

```python
class Call:   
    def __init__(self, P_0, X, mu, sigma, T, seed):        
        ### Create an instance of the Call class with its own data
       
        self.P_0 = P_0
        self.X = X
        self.mu = mu
        self.sigma = sigma
        self.T = T
        self.seed = seed
               
    def monte_carlo(self):        
        #### Create a simualted price path based on the instance's data
        
        self.P = np.zeros(self.T)
        self.P[0] = self.P_0
        
        # Set random number seed
        np.random.seed(self.seed)
        
        # Generate random numbers
        self.Z = np.random.normal(self.mu, self.sigma, self.T)
        
        # Convert random numbers to multipliers
        self.R = self.Z + 1
        
        # Use a For loop to fill in prices
        for i in range(1, self.T):
            self.P[i] = self.P[i-1] * self.R[i-1]
            
    def plot_path(self):       
        ### Quick plot of the price path
        
        self.x_values = np.linspace(0, self.T  - 1, self.T)
        plt.plot(self.x_values, self.P)    
                        
    def compute_payout(self):        
        # Payout max(0, P[-1] - X)
        
        self.payout = max(0, self.P[-1] - self.X)
        
        print('The price path finished with a price of ' + str(self.P[-1]))
        print('The payout is ' + str(self.payout))
```
